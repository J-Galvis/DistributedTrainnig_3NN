### Define functions for comparassion

```python
def main(folder: str = "../stats/CIFAR_10"):

    runs = load_training_folder(folder)  

    print(runs_metadata(runs).to_string())

    chart = compare_runs(runs, loose=True)
    chart.display()
    return 0

def speedup_analysis(folder: str = "../stats/CIFAR_10", base_case: str = None):

    runs = load_training_folder(folder)
    
    # Display metadata
    print(f"Speedup comparison (baseline: '{base_case}'):")
    print(runs_metadata(runs).to_string())
    
    # Create speedup visualization
    chart = compare_speedups(runs, base_case=base_case)
    chart.display()
    
    return runs
```
